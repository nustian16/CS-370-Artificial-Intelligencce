```python
from PIL import Image, ImageDraw
img = Image.open("test.jpg").convert("1")
#img.show()
```

### Below is the picture of signature I intend to use to complete all the lab tasks. For testing purpose I have drawn a border around the image.

<img src="test1.png">

```python
left, top = img.size
width, height = img.size
right, bottom = 0,0
print(img.size)
```

```python
for (x,y) in [(x,y) for x in range(width) for y in range(5,height-5)]:
    color = img.getpixel((x, y))
    if color is 0:
        if x > right:
            right = x
        if x < left:
            left = x
        if y > bottom:
            bottom = y
        if y < top:
            top = y
            
print(left,right,top,bottom)
draw = ImageDraw.Draw(img)
draw.rectangle(((0, 0), (img.size[0]-1,img.size[1]-1)))
draw.rectangle(((left, top), (right, bottom)))
#img.show()
```

### The above algorithm was the first lab task. This algorithm basically finds the cordinates of the bounding box of the signature. The result of above algorithm is displayed in the diagram below:

<img src="test2.png">

```python
cx = 0
cy = 0
n = 0
for (x,y) in [(x,y) for x in range(0,width) for y in range(0,height)]:
    color = img.getpixel((x, y))
    if color is 0:
        cx = cx + x
        cy = cy + y
        n = n + 1
cx = cx // n
cy = cy // n
print(cx,cy)
```

### The above algorithm basically finds the centroid for the given signature. This was the second lab task. The output of the above cell is shown right after the cell.

```python
seg1 = (left, cx, top, cy) 
seg2 = (cx, right, top, cy) 
seg3 = (left, cx, cy, bottom) 
seg4 = (cx, right, cy, bottom)
seg_list = seg1,seg2,seg3,seg4
for segs in seg_list:
    draw.rectangle(((segs[0], segs[2]), (segs[1], segs[3])))
print(seg1)
print(seg2)
print(seg3)
print(seg4)
#img.show()
```
